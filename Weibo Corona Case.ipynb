## Scraping User Profile
Now we have the user profile IDs, let's get their profile info on Weibo. After some research, it seems like their mobile page is easier to scrape.

First, we will need our cookie information once we logged in on Weibo.(https://m.weibo.cn/) We can get this information using Chrome's DevTools:


Copy this information and save it to a CSV file, in this case, I'll use "cookie.csv" as filename. To use this cookie:

```python
cookie_data = pd.read_csv('./cookie.csv', sep=' ', index_col=False)
lis = cookie_data.values.tolist()
cookie = {}
for sep in lis:
    cls = sep[0].split('\t')
    cookie.setdefault(cls[0], cls[1])
```

Also we need to define user agents and header for the request:

```python
import random
user_agents = [
    "Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36",
    "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.153 Safari/537.36",
    "Mozilla/5.0 (Windows NT 6.1; WOW64; rv:30.0) Gecko/20100101 Firefox/30.0"
    "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.75.14 (KHTML, like Gecko) Version/7.0.3 Safari/537.75.14",
    "Mozilla/5.0 (compatible; MSIE 10.0; Windows NT 6.2; Win64; x64; Trident/6.0)"
    ]


header = {
    "MWeibo-Pwa": "1",
    "X-XSRF-TOKEN": "fffd26",
    "Referer": "https://m.weibo.cn/p/index?containerid=2302832281700573_-_INFO&title=%E5%9F%BA%E6%9C%AC%E8%B5%84%E6%96%99&luicode=10000011&lfid=2302832281700573&sudaref=m.weibo.cn",
    "User-Agent": random.choice(user_agents),
    'Accept': 'application/json, text/plain, */*'
}

random_seconds = [1, 2, 3, 4]
```

Now let's define a function to get their container ID and later use it to get their profile page:

```python
import requests
import json
import time
def get_container_id(user_id):
    # This is the API url after analyzing their page
    container_url = 'https://m.weibo.cn/api/container/getIndex?type=uid&value=' + user_id
    res = requests.get(url=container_url, headers=header, verify=False, cookies=cookie)
    content = json.loads(res.content).get('data')

    # Stop for a few seconds between each request
    time.sleep(random.choice(random_seconds))
    container_id = None
    if content.get('tabsInfo') is None:
        return container_id
#     print(content)
    for data in content.get('tabsInfo').get('tabs'):
        if (data.get('tab_type') == 'profile'):
            container_id = data.get('containerid')
    return container_id
```

Now we can use this ID to request their profile info, but first, let's make a class to store them:

```python
class User:
    def __init__(self, nickname, desc, register_time, credit, sex, birthday, address, email, phone):
        self.nickname = nickname
        self.desc = desc
        self.register_time = register_time
        self.credit = credit
        self.sex = sex
        self.birthday = birthday
        self.address = address
        self.email = email
        self.phone = phone

    def to_list(self):
        return [
            self.nickname,
            self.desc,
            self.register_time,
            self.credit,
            self.sex,
            self.birthday,
            self.address,
            self.email,
            self.phone
        ]
```

Now define a function to get their info:

```python
from urllib.parse import urlencode
def get_user_info(container_id):
    info = {'title': '基本资料'}
    info = urlencode(info)

    # API to get their profile info
    url = 'https://m.weibo.cn/api/container/getIndex?containerid={0}_-_INFO&{1}&luicode=10000011&lfid={0}'.format(
        container_id, info)

    res = requests.get(url=url, verify=False, headers=header, cookies=cookie)
    time.sleep(random.choice(random_seconds))
    json_user = res.json()
    cards_list = json_user['data']['cards']

    # Initialize an empty User to store data
    user_info = User('', '', '', '', '', '', '', '', '',)
    for card_group in cards_list:
        for card in card_group.keys():
            obj = card_group[card]

            # ['昵称', '简介', '注册时间', '阳光信用', '性别', '生日', '所在地', '邮箱', '电话'],
            if type(obj) is list:
                for info in obj:
#                     print(info)
                    if str(info).find('昵称') > -1:
                        user_info.nickname = info['item_content']

                    if str(info).find('简介') > -1:
                        user_info.desc = info['item_content']

                    if str(info).find('注册时间') > -1:
                        user_info.register_time = info['item_content']

                    if str(info).find('阳光信用') > -1:
                        user_info.credit = info['item_content']

                    if str(info).find('性别') > -1:
                        user_info.sex = info['item_content']

                    if str(info).find('生日') > -1:
                        user_info.birthday = info['item_content']
                        # There are formats like "10-03  天秤座"  therefore cannot calculate age
                        """
                        birth = info['item_content'] + " 00:00:00"
                        birth_date = datetime.datetime.strptime(birth, '%Y-%m-%d')
                        age = calculate_age(birth_date)
                        user_info.setdefault('age', age)
                        """

                    if str(info).find('所在地') > -1:
                        user_info.address = info['item_content']

                    if str(info).find('邮箱') > -1:
                        user_info.email = info['item_content']

                    if str(info).find('电话') > -1:
                        user_info.telephone = info['item_content']
    return user_info
```

Now let's get the first 20 user profiles:

```python
user_list = []
error_list = []

for i, id in enumerate(ids):
    print(f"Start scraping {i + 1} in {len(ids)}, ID: {id}")
    try:
        container_id = get_container_id(id)
    except Exception as e:
        print("ID: " + id + ", Error:")
        print(e)
        error_list.append(id)
        continue
    if container_id is None:
        print(id + ': User page empty')
        continue
    try:
        user = get_user_info(container_id)
        user_list.append(list(user.to_list()))
    except Exception as e:
        print("ID: " + id + ", Error:")
        print(e)
        error_list.append(id)
        continue
if error_list:
    print("\n\n IDs with error：", error_list)
```

Now we can store this data to an Excel file so it can be analyzed later:

```python
import os
from datetime import datetime

def get_full_file_name(base='./'):
    if not os.path.exists(base):
        os.mkdir(base)

    now = datetime.now()
    folder = now.strftime('%Y%m%d')

    if not os.path.exists(base + folder):
        os.mkdir(base + folder)

    suffix = datetime.now().strftime('%Y%m%d%H%M%S')
    return base + folder + '/', suffix
```

```python
import numpy as np

columns_names = ['昵称', '简介', '注册时间', '阳光信用', '性别', '生日', '所在地', '邮箱', '电话']
if user_list:
    data_frame = pd.DataFrame(data=np.array(user_list), columns=columns_names)
    folder, suffix = get_full_file_name()
    excel_name = folder + '/res_{0}.xlsx'.format(suffix)
    data_frame.to_excel(excel_name)
```

## Data Visualization

Here are some example to show data in charts:

```python
import matplotlib.pyplot as plt
from matplotlib import pylab
import warnings

def show_sex_bar_chart(pd):
    # Display Chinese
    pylab.mpl.rcParams['font.sans-serif'] = ['SimHei']

    # Disable warning
    warnings.filterwarnings("ignore", category=plt.cbook.mplDeprecation)

    # Count different sex
    df = pd['性别'].value_counts()

    print(df)
    name_list = list(df.index)
    for i, name in enumerate(name_list):
        if name is None or len(name) == 0:
            name = '未知'
            name_list[i] = name

    num_list = df.values.tolist()

    # Display bar chart
    plt.bar(range(len(num_list)), num_list, color='rgb', tick_label=name_list)
    # plt.savefig(os.path.abspath(os.curdir) + os.sep + str(random.randint(1, 10000)))
    plt.show()
```
